# Predicting Fault Location 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
df = pd.read_csv("ccms_data.csv")   # Using data generated randomly
```

```python
df.head()
```

### Creating Feature Matrix

```python
X = df.drop("distance",axis=1)
```

### Creating Target Vetor

```python
y = df.iloc[:,-1]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
print(type(y_train))
```

### Creating Classes

```python
from keras.utils import np_utils
num_classes = len(np.unique(y_train))
```

```python
num_classes
```

```python
y_train_new = []
for val in y_train:
    y_train_new.append(val//30-1)
```

```python
y_test_new = []
for val in y_test:
    y_test_new.append(val//30-1)
```

### Encoding of Target Value

```python
y_train = np_utils.to_categorical(y_train_new,num_classes)
y_test = np_utils.to_categorical(y_test_new,num_classes)
```

```python
#X_train,X_val = X_train[:],X_train[:]
#X_train,X_val = X_train[:],X_train[:]
```

```python
print(X_train.iloc[0].shape)
#print(X_test[0].shape)
#print(X_val[0].shape)
```

### Creating Model

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten
model = Sequential()
model.add(Dense(120, input_dim=8 , activation='softmax'))
model.add(Dense(70, activation='softmax'))
model.add(Dense(30, activation='softmax'))
model.summary()
```

### Compiling the model

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', 
                  metrics=['accuracy'])
```

### Fitting the data

```python
model.fit(X_train,y_train,epochs=500,batch_size=100)
```
